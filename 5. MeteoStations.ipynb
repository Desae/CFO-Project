**_First request Returns a list of all meteostations for which data is available._**

The keys in the requests are:
1.Type
2.Features
3.Geometry -(type, coordinates)
4.Properties -(meteostationid, name, longitude, latitude, altitude, source,wmocode, provider)

```python
def getData(url):
    data = requests.get(url, headers=headers)
    return data.json()
```

```python
url = "http://agrodatacube.wur.nl/api/v1/rest/meteostations?output_epsg=28992&page_size=25&page_offset=0"
getData(url)
```
