### Description

Requests to retrieve data from the Dutch soil map 1:50.000 (2014). This dataset is provided by WUR, and also avaible as Open Data. For more information please see https://www.wur.nl/nl/show/Bodemkaart-1-50-000.htm.

**_First request Returns the intersections of the supplied geometry and soilmap. If no epsg is provided epsg = 28992 (RD) is assumed. Currently only epsg's 28992 and 4326 (WGS 84) are supported_**

```python
def getData(url):
    data = requests.get(url, headers=headers)
    return data.json()
```

```python
url = "http://agrodatacube.wur.nl/api/v1/rest/soiltypes?page_size=1000&page_offset=1"
getData(url)
```

**_Second request Return all the information for the given soilmap entityid._**

```python
url = "http://agrodatacube.wur.nl/api/v1/rest/soiltypes/8?output_epsg=28992&page_size=25&page_offset=0"
getData(url)
```
