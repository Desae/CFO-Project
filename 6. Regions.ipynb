**_First request Returns all municipalities (gemeentes), based on the boundaries for 2015._**

```python
def getData(url):
    data = requests.get(url, headers=headers)
    return data.json()
```

```python
url = "http://agrodatacube.wur.nl/api/v1/rest/regions/municipalities?page_size=25&page_offset=0&output_epsg=28992"
getData(url)
```

**_Second request Return the dutch 6-positions postalcode areas._**

```python
url = "http://agrodatacube.wur.nl/api/v1/rest/regions/postalcodes?page_size=25&output_epsg=28992&page_offset=0"
getData(url)
```

**_Third request Return the provinces. These are the boundaries from 2015._**
